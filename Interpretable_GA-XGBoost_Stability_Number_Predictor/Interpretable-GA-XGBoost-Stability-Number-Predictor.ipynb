## Modeling

```python
# Train_Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3) 
```

```python
# Import Models 
import xgboost as xgb
xgb_regressor = xgb.XGBRegressor()

from sklearn.model_selection import KFold
```

```python
# Hyperparameters of XGBoost model

# max depth
max_depth = [int(x) for x in np.linspace(1, 15, num = 15)]
    
# number of estimators 
n_estimators = [int(x) for x in np.linspace(1000, 3000, num = 21)]

# learning rate
learning_rate = [float(x) for x in np.linspace(0.001, 1, num = 21)]

        

params_grid={"max_depth": max_depth,
           "n_estimators":n_estimators,
           "learning_rate":learning_rate,
           "lambda": np.arange(0,11,1) ,
           "gamma": np.arange(0,11,1)}

print(params_grid)
```

```python
# hyperameter tunining using evolutionary search algoprithm
from evolutionary_search import EvolutionaryAlgorithmSearchCV
cv = EvolutionaryAlgorithmSearchCV(estimator=xgb_regressor,
                                   params=params_grid,
                                   cv=10,
                                   verbose=1,
                                   population_size=50,
                                   gene_mutation_prob=0.010,
                                   gene_crossover_prob=0.7,  
                                   tournament_size=3,
                                   generations_number=20,
                                   n_jobs=1)
cv.fit(X_train, y_train)
```

```python
# Model details 
dir(cv)
```

## Best model

```python
xgb_regressor = xgb.XGBRegressor (max_depth= 5, n_estimators= 1900, learning_rate = 0.75025, reg_lambda=8, gamma=1)
xgb_regressor.fit(X_train, y_train)
```

```python
xgb_regressor.score(X_test, y_test)
```

# Model Evaluation 

# 1. Training Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_train_pred = xgb_regressor.predict(X_train)
```

```python
mae = mean_absolute_error(y_train, y_train_pred)
mse = mean_squared_error(y_train,y_train_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_train, y_train_pred)
```

```python
print ('MAE:', mae)
print ('MSE:', mse)
print ('RMSE:', rmse)
print ('R-Squared:', r2)
```

# 2. Testing Evaluation


```python
y_pred = xgb_regressor.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print ('MAE:', mae)
print ('MSE:', mse)
print ('RMSE:', rmse)
print ('R_sqaured:', r2)
```
